```python
for i in tqdm(range(len(train_prob))):
    ids = list(labeled_train_err[(labeled_train_err.user_id == train_prob.user_id[i]) \
                           & (labeled_train_err.time < train_prob.time[i])].index)
    labeled_train_err.target[ids] = 1
```

```python
pd.set_option('mode.chained_assignment',  None) # 경고 무시
def labeling(err_df, prob_df, days=1): # days : 불만 제기 이전에 마주한 오류 범위 지정(날짜), (2 : 2일전까지, 3 : 3일전까지)
    labeled_df = err_df.copy()
    targets = np.zeros(len(labeled_df), dtype=int)
    labeled_df['target'] = targets
    
    for i in tqdm(range(len(train_prob))):
        labeled_df[(labeled_df.user_id == prob_df.user_id[i])\ 
                   & (labeled_df.time < prob_df.time[i] - dt.timedelta(days=days-1))].target = 1
    
    return labeled_df
```
